```python
def decode_varint(pos, data):
     shift = 0
     result = 0
     while True:
         i = data[pos]
         pos += 1
         result |= (i & 0x7f) << shift
         shift += 7
         if not (i & 0x80):
             break
     return pos, result

def decode_zigzag(n):
     return (n >> 1) ^ (-(n & 1))
```

```python
def dum_dat(dtype):
    print(dtype)
    if dtype['type'] == 'int':
        return "np.int32(0)"
    if dtype['type'] == 'long':
        return "np.int64(0)"
    if dtype['type'] == 'float':
        return "np.float32(0)"
    if dtype['type'] == 'double':
        return "np.float64(0)"
    if dtype['type'] == 'boolean':
        return "0"
    if dtype['type'] == 'bytes':
        return "b'a'"
```

```python
def rec_exp_json_code(file,fields,pos,strr,ind,aform,count,dec):
    if file["type"] == "null":
        return aform,strr,count,dec
    
    elif file["type"] == "record":
        #print(file["type"],count)
        temp = count
        aform.append( '{"class" : "RecordArray",')
        aform.append('"contents": {\n')
        for elem in file["fields"]:
            aform.append('"'+elem["name"]+'"'+": ")
            aform,strr,count,dec = rec_exp_json_code(elem,fields,pos,strr,ind,aform,count+1,dec)
        aform.append( '}},"form_key": "node{}"}},\n'.format(temp))
        return aform,strr,count,dec
    
    elif file["type"] == "string":
        #print(file["name"])
        #print(file["type"],count)
        aform.append('{{"class": "ListOffsetArray64","offsets": "i64","content": {{"class": "NumpyArray","primitive": "uint8","parameters": {{"__array__": "char"}},"form_key": "node{}"}},"parameters": {{"__array__": "string"}},"form_key": "node{}"}},\n'.format(count+1,count))
        var1 = " 'part0-node{}-offsets'".format(count)
        var2 = " 'part0-node{}-data'".format(count+1)
        dec.append(var1)
        dec.append(": [0],")
        dec.append(var2)
        dec.append(": [],")
        strr.append("\n"+"    "*ind+"pos, inn = decode_varint(pos,fields)")
        strr.append("\n"+"    "*ind+"out = decode_zigzag(inn)")
        strr.append("\n"+"    "*ind+"con['part0-node{}-offsets'].append(np.uint8(out+con['part0-node{}-offsets'][-1]))".format(count,count))
        #print(pos,abs(out))
        strr.append("\n"+"    "*ind+"con['part0-node{}-data'].append(fields[pos:pos+abs(out)].tobytes().decode(errors=\"surrogateescape\"))".format(count+1))
        strr.append("\n"+"    "*ind+"print(fields[pos:pos+abs(out)].tobytes().decode(errors=\"surrogateescape\") )")
        strr.append("\n"+"    "*ind+"pos = pos + abs(out)")
        return aform,strr,count+1,dec
    
    elif file["type"] == "int":
        #print(file["name"])
        #print(file["type"],count)
        aform.append('{{"class" : "NumpyArray", "primitive": "int32", "form_key": "node{}"}},\n'.format(count))
        var1 = " 'part0-node{}-data'".format(count)
        dec.append(var1)
        dec.append(": [],")
        strr.append("\n"+"    "*ind+"pos, inn = decode_varint(pos,fields)")
        strr.append("\n"+"    "*ind+"out = decode_zigzag(inn)")
        strr.append("\n"+"    "*ind+"con['part0-node{}-data'].append(np.int32(out))".format(count))
        strr.append("\n"+"    "*ind+"print(out)")
        return aform,strr,count,dec
    
    elif file["type"] == "long":
        #print(file["name"])
        #print(file["type"],count)
        aform.append('{{"class" : "NumpyArray", "primitive": "int64", "form_key": "node{}"}},\n'.format(count))
        var1 = " 'part0-node{}-data'".format(count)
        dec.append(var1)
        dec.append(": [],")
        strr.append("\n"+"    "*ind+"pos, inn = decode_varint(pos,fields)")
        strr.append("\n"+"    "*ind+"out = decode_zigzag(inn)")
        strr.append("\n"+"    "*ind+"print(out)")
        strr.append("\n"+"    "*ind+"con['part0-node{}-data'].append(np.int64(out))".format(count))
        return aform,strr,count,dec
    
    elif file["type"] == "float":
        #print(file["name"])
        #print(file["type"],count)
        aform.append('{{"class" : "NumpyArray", "primitive": "float32", "form_key": "node{}"}},\n'.format(count))
        var1 = " 'part0-node{}-data'".format(count)
        dec.append(var1)
        dec.append(": [],")
        strr.append("\n"+"    "*ind+"print(struct.Struct(\"<f\").unpack(fields[pos:pos+4].tobytes())[0])")
        strr.append("\n"+"    "*ind+"con['part0-node{}-data'].append(np.float32(struct.Struct(\"<f\").unpack(fields[pos:pos+4].tobytes())[0]))".format(count))
        strr.append("\n"+"    "*ind+"pos = pos+4")
        return aform,strr,count,dec
    
    elif file["type"] == "double":
        #print(file["name"])
        #print(file["type"],count)
        aform.append('{{"class" : "NumpyArray", "primitive": "float64", "form_key": "node{}"}},\n'.format(count))
        var1 = " 'part0-node{}-data'".format(count)
        dec.append(var1)
        dec.append(": [],")
        strr.append("\n"+"    "*ind+"print(struct.Struct(\"<d\").unpack(fields[pos:pos+8].tobytes())[0])")
        strr.append("\n"+"    "*ind+"con['part0-node{}-data'].append(np.float64(struct.Struct(\"<d\").unpack(fields[pos:pos+8].tobytes())[0]))".format(count))        
        strr.append("\n"+"    "*ind+"pos = pos+8")
        return aform,strr,count,dec
    
    elif file["type"] == "boolean":
        #print(file["name"])
        #print(file["type"],count)
        aform.append('{{"class" : "NumpyArray", "primitive": "bool", "form_key": "node{}"}},\n'.format(count))
        var1 = " 'part0-node{}-data'".format(count)
        dec.append(var1)
        dec.append(": [],")
        strr.append("\n"+"    "*ind+"print(fields[pos])")
        strr.append("\n"+"    "*ind+"con['part0-node{}-data'].append(fields[pos])".format(count))
        strr.append("\n"+"    "*ind+"pos = pos+1")
        return aform,strr,count,dec
    
    elif file["type"] == "bytes":
        #print(file["name"])
        #print(file["type"],count)
        astring = '{{"class": "ListOffsetArray64","offsets": "i64","content": {{"class": "NumpyArray", "primitive": "uint8","parameters": {{"__array__": "char"}},"form_key": "node{}"}},"parameters": {{"__array__": "string"}},"form_key": "node{}"}},\n'.format(count+1,count)
        var1 = " 'part0-node{}-data'".format(count+1)
        var2 = " 'part0-node{}-offsets'".format(count)
        dec.append(var2)
        dec.append(": [0],")
        dec.append(var1)
        dec.append(": [],")
        aform .append(astring)
        strr.append("\n"+"    "*ind+"pos, inn = decode_varint(pos,fields)")
        strr.append("\n"+"    "*ind+"out = decode_zigzag(inn)")
        strr.append("\n"+"    "*ind+"con['part0-node{}-offsets'].append(out+con['part0-node{}-offsets'][-1])".format(count,count))
        strr.append("\n"+"    "*ind+"print(fields[pos:pos+out].tobytes())")
        strr.append("\n"+"    "*ind+"con['part0-node{}-data'].append(fields[pos:pos+out].tobytes())".format(count+1))
        strr.append("\n"+"    "*ind+"pos = pos+out")
        return aform,strr,count+1,dec
    
    elif isinstance(file["type"],list):
        #print(file["name"])
        temp = count
        strr.append("\n"+"    "*ind+"pos, inn = decode_varint(pos,fields)")
        strr.append("\n"+"    "*ind+"out = abs(decode_zigzag(inn))")
        strr.append("\n"+"    "*ind+"print(\"index :\",out)")
        out = len(file["type"])
        if out == 2:
            if "null" in file["type"]:
                aform.append('{"class": "ByteMaskedArray","mask": "i8","content":\n')
                var1 = " 'part0-node{}-mask'".format(count)
                dec.append(var1)
                dec.append(": [],")
        temp = count
        idxx = file['type'].index('null')
        for i in range(out):
            if file["type"][i] == "null":
                strr.append("\n"+"    "*(ind)+"if out == {}:".format(i))
                strr.append("\n"+"    "*(ind+1)+"con['part0-node{}-mask'].append(np.int8(False))".format(temp))
                strr.append("\n"+"    "*(ind+1)+"con['part0-node{}-data'].append({})".format(temp+1,dum_dat({'type': file['type'][1-idxx]})))
            else:
                strr.append("\n"+"    "*(ind)+"if out == {}:".format(i))
                strr.append("\n"+"    "*(ind+1)+"con['part0-node{}-mask'].append(np.int8(True))".format(count))
                aform,strr,count,dec = rec_exp_json_code({"type": file["type"][i]},fields,pos,strr,ind+1, aform,count+1,dec)
        aform.append('"valid_when": true,"form_key": "node{}"}}\n'.format(temp))
        return aform,strr,count,dec
    
    elif isinstance(file["type"],dict):
        #print(file["name"])
        #print(file["type"],count)
        aform,strr,count,dec = rec_exp_json_code(file["type"],fields,pos,strr,ind, aform,count,dec)
        return aform,strr,count,dec
    
    elif file["type"] == "fixed":
        #print(file["type"],count)
        var1 = " 'part0-node{}-data'".format(count+1)
        dec.append(var1)
        dec.append(": [],")
        aform.append('{{"class": "RegularArray","content": {{"class": "NumpyArray","primitive": "uint8","form_key": "node{}"}},"size": {},"form_key": "node{}"}},\n'.format(count+1,file["size"],count))
        strr.append("\n"+"    "*ind+"lenn = {}".format(file["size"]))
        strr.append("\n"+"    "*ind+"print(fields[pos:pos+lenn].tobytes())")
        strr.append("\n"+"    "*ind+"for i in range(lenn):")
        strr.append("\n"+"    "*(ind+1)+"con['part0-node{}-data'].append(np.uint8(fields[i]))".format(count+1))
        strr.append("\n"+"    "*ind+"pos = pos+lenn")
        return aform,strr,count+1,dec
    
    elif file["type"] == "enum":
        #print(file["name"])
        #print(file["type"],count)
        aform.append('{{"class": "IndexedArray64","index": "i64","content": {{"class": "ListOffsetArray64","offsets": "i64","content": {{"class": "NumpyArray","primitive": "uint8","parameters": {{"__array__": "char"}},"form_key": "node{}"}},"parameters": {{"__array__": "string"}},"form_key": "node{}"}}, "parameters": {{"__array__": "categorical"}}, "form_key": "node{}"}},\n'.format(count+2,count+1,count))
        var1 = " 'part0-node{}-index'".format(count)
        dec.append(var1)
        dec.append(": [],")
        tempar = file['symbols']
        offset, dat = [0] , []
        prev = 0
        for i in range(len(tempar)):
            offset.append(len(tempar[i])+prev)
            prev = offset[-1]
        for i in range(len(tempar)):
            for elem in range(len(tempar[i])):
                dat.append(elem)
        var2 = " 'part0-node{}-offsets': {},".format(count+1,str(offset))
        dec.append(var2)
        var2 = " 'part0-node{}-data': {},".format(count+2,str(dat))
        dec.append(var2)
        strr.append("\n"+"    "*ind+"pos, inn = decode_varint(pos,fields)")
        strr.append("\n"+"    "*ind+"out = decode_zigzag(inn)")
        strr.append("\n"+"    "*ind+"con['part0-node{}-index'].append(np.int64(out))".format(count))
        strr.append("\n"+"    "*ind+"print(out)")
        return aform,strr,count+2,dec
    
    elif file["type"] == "array":
        #print(file["name"])
        temp = count
        var1 = " 'part0-node{}-offsets'".format(count)
        dec.append(var1)
        dec.append(": [0],")
        aform.append('{"class": "ListOffsetArray64","offsets": "i64","content": ')
        strr.append("\n"+"    "*ind+"pos, inn = decode_varint(pos,fields)")
        strr.append("\n"+"    "*ind+"out = decode_zigzag(inn)")
        strr.append("\n"+"    "*ind+"print(\"length\",out)")
        strr.append("\n"+"    "*ind+"con['part0-node{}-offsets'].append(out+con['part0-node{}-offsets'][-1])".format(count,count))
        strr.append("\n"+"    "*ind+"for i in range(out):")
        aform,strr,count,dec = rec_exp_json_code({"type": file["items"]},fields,pos,strr,ind+1,aform,count+1,dec)
        strr.append("\n"+"    "*ind+"pos, inn = decode_varint(pos,fields)")
        strr.append("\n"+"    "*ind+"out = decode_zigzag(inn)")
        strr.append("\n"+"    "*ind+'''if out != 0:''')
        strr.append("\n"+"    "*(ind+1)+"raise")
        aform.append( '"form_key": "node{}"}},\n'.format(temp))
        return aform,strr,count,dec
    
    elif file["type"] == "map":
        #print(file["name"])
#         strr.append("\npos, inn = decode_varint(pos,fields)"
#         strr.append("\nout = abs(decode_zigzag(inn))"
#         strr.append("\nprint(\"length\",out)"
#         strr.append("\nfor i in range(out):"
#         strr.append("\n"+"    "*(ind+1)+"print(\"{{\")"
#         strr = strr+aa
#         strr = strr+bb
#         strr = strr+ccfa
#         strr = strr+dd
#         strr = strr+ee
#         pos,strr,count = rec_exp_json_code({"type": file["values"]},fields,pos,strr,ind+1, aform,count+1)
#         strr.append("\n    "*(ind+1)+"print(\":\")"
#         strr = strr+ff
#         pos,strr,count = rec_exp_json_code({"type": file["values"]},fields,pos,strr,ind+1, aform,count+1)
#         strr.append("\n"+"    "*(ind+1)+"print(\"}}\")"
#         strr.append("\n"+"    "*ind+"pos, inn = decode_varint(pos,fields)"
#         jj = "\n"+"    "*ind+"out = decode_zigzag(inn)"
#         kk = "\n"+"    "*ind+'''if out != 0:
#     raise
#             '''
#         strr = strr+gg
#         strr = strr+hh
#         strr = strr+jj
#         strr = strr+kk
        raise NotImplementedError
#        return aform,strr,count
```
